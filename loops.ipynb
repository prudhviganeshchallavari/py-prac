##### The Basic Counter
Write a while loop that prints the numbers from 10 down to 1. After the loop finishes, it should print "Liftoff!".

```python
i=10
while i>=1:
    print(i,end=' ')
    i=i-1
```

##### Use a for loop and the range() function to calculate the sum of all even numbers between 1 and 50 (inclusive). Print the final total.

```python
for num in range(1,51):
    if num%2==0:
        print(num,end=' ')
```

##### Ask the user for a number. Use a loop to print the multiplication table for that number from 1 to 10 (e.g., "5 x 1 = 5", "5 x 2 = 10", etc.).

```python
n = int(input('enter a number:'))
for i in range(1,11):
    print(n*i,end=' ')
```

#### List Filter
Given a list of names: names = ["Alice", "Bob", "Charlie", "David", "Eve", "Frank"].
Use a loop to print only the names that have more than 4 characters.

```python
names = ["Alice", "Bob", "Charlie", "David", "Eve", "Frank"]
for name in names:
    if len(name)<4:
        print(name,sep=';')
```

##### The Password Guard: Create a while True loop that keeps asking the user for a password. If they type "secret123", print "Access Granted" and break. Otherwise, keep asking.

```python
password = 'secret123' 
user_pass = input('enter your password:')
while True:
    if password==user_pass:
        print('access granted')
        break
    else:
        print('access denied')
        break
```
