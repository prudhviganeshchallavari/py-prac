#### Write a program that asks the user for a number. If the number is even, print "This is an even number." If it's odd, print "This is an odd number."

##### Create a variable score. If the score is 50 or greater, print "Pass". Otherwise, print "Fail".

```python
##1st solution
num = int(input('enter a num:'))
if num%2==0:
    print('even')

else:
    print('odd')
```

```python
##2nd solution
score = int(input('enter score:'))
if score>=50:
    print('pass')

else:
    print('fail')
```

### Write a script where a user inputs a numerical grade (0-100), and the program prints the letter grade based on these rules:

90-100: A

80-89: B

70-79: C

60-69: D

Below 60: F

```python
##3rd solution
marks = int(input('enter your marks:'))
if marks<=100 and marks>=90:
    print('A')

elif marks<=89 and marks>=80:
    print('B')

elif marks<=79 and marks>=70:
    print('C')

elif marks<=69 and marks>=60:
    print('D')

else:
    print('FAILED!!!')
```
