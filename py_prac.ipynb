### Write a script where a user inputs a numerical grade (0-100), and the program prints the letter grade based on these rules:

90-100: A

80-89: B

70-79: C

60-69: D

Below 60: F

```python
##3rd solution
marks = int(input('enter your marks:'))
if marks<=100 and marks>=90:
    print('A')

elif marks<=89 and marks>=80:
    print('B')

elif marks<=79 and marks>=70:
    print('C')

elif marks<=69 and marks>=60:
    print('D')

else:
    print('FAILED!!!')
```

##### Write a program to check whether a year is a leap year or not.

```python
year = int(input('enter a year:'))
if year%4==0 and year%400==0:
    print('its leap year')

elif year%100!=0:
    print('its a leap')

else:
    print('not a leap year')
```

##### Create a simple calculator using if-elif-else for +, -, *, /.

```python
num1 = int(input('enter a number:'))
num2 = int(input('enter a number:'))
operator = input('enter a operator:(+,-,/,*)')
if operator == '+':
    print(num1+num2)

elif operator=='-':
    print(num1-num2)

elif operator=='*':
    print(num1*num2)

elif operator=='/':
    print(num1/num2)

else:
    print('enter a valid operator')
```

##### Write a program to check if a word is a palindrome.

```python
word = input('enter a word:')
if word==word[::-1]:
    print('its a palindrome')

else:
    print('not a palindrome')
```

#### Create a login system:

If username == "admin" and password == "1234", print "Access granted"

Otherwise, print "Access denied"

```python
user = 'admin' 
password = '1234' 
my_user = input('enter user:')
my_password = input('enter password:')
if my_user==user and my_password==password:
    print('access granted')
elif my_user==user and my_password!=password:
    print('incorrect details')
elif my_user!=user and my_password==password:
    print('incorrect details')
else:
    print('access denied')
```

### Create a password strength checker:

Length < 6 → Weak

6–10 → Medium

10 → Strong

```python
password = input()
if len(password)<6:
    print('weak password')

elif len(password)>=6 and len(password)<10:
    print('medium password')

elif password>=10:
    print('strong password')

else:
    print('atleast enter credentials')
```
